```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt

plt.style.use('dark_background')
```

```python
import tensorflow as tf

(X_train, y_train), (X_test, y_test) = tf.keras.datasets.fashion_mnist.load_data()
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
X_train = X_train[:10000]
X_test = X_test[:10000]

y_train = y_train[:10000]
y_test = y_test[:10000]
```

```python
plt.figure(figsize=(16, 6))
for i in range(10):
    plt.subplot(2, 5, i + 1)
    plt.imshow(X_train[i, :], cmap='gray')
```

```python
X_train = X_train.reshape(X_train.shape[0], -1)
X_test = X_test.reshape(X_test.shape[0], -1)

X_train.shape
```

```python
# scaling is not necessary!
```

#### Model training

```python
%%time

model = DecisionTreeClassifier(random_state=1)
model.fit(X_train, y_train)
```

```python
pred_train = model.predict(X_train)
pred_train
```

```python
train_preds = pd.DataFrame(data=pred_train, columns=['pred'])
train_preds['true'] = y_train

train_preds
```

```python
pred_test = model.predict(X_test)
```

```python
test_preds = pd.DataFrame(data=pred_test, columns=['preds'])
test_preds['true'] = y_test
test_preds
```

#### Metric: accuracy

This metric good work with balanced data

```python
train_preds['true'].value_counts()
```

```python
test_preds['true'].value_counts()
```

```python
print(f'Train accuracy: {accuracy_score(y_train, pred_train)}')
print(f'Test accuracy: {accuracy_score(y_test, pred_test)}')
```

Let's improve the result

```python
from sklearn.tree import plot_tree

plt.figure(figsize=(15, 15))
plot_tree(model, max_depth=1)
plt.show()
```

```python
model = DecisionTreeClassifier(random_state=1, min_samples_leaf=2, max_depth=10)
model.fit(X_train, y_train)
train_pred = model.predict(X_train)
test_pred = model.predict(X_test)
```

```python
train_preds = pd.DataFrame(data=train_pred, columns=['preds'])
train_preds['true'] = y_train

train_preds
```

```python
test_preds = pd.DataFrame(data=test_pred, columns=['preds'])
test_preds['true'] = y_test

test_preds
```

#### Validation

```python
print(f'Train accuracy: {accuracy_score(y_train, pred_train)}')
print(f'Test accuracy: {accuracy_score(y_test, pred_test)}')
```

```python
print(f'Train accuracy: {accuracy_score(y_train, train_pred)}')
print(f'Test accuracy: {accuracy_score(y_test, test_pred)}')
```

```python
from sklearn.model_selection import GridSearchCV

param = {'min_samples_leaf': [1, 3, 5, 7, 10],
         'max_depth': [3, 5, 8, 15]}

gs = GridSearchCV(model, param, n_jobs=-1)
gs.fit(X_train, y_train)
```
