```python
import pandas as pd
import numpy as np

from sklearn.datasets import load_iris
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import f1_score

import matplotlib.pyplot as plt
```

```python
iris_data = load_iris()
```

```python
colors = ['blue', 'red', 'green']
for feature in range(iris_data.data.shape[1]):
    plt.subplot(2, 2, feature+1)
    for label, color in zip(range(len(iris_data.target_names)), colors):
        # find the label and plot the corresponding data
        plt.hist(iris_data.data[iris_data.target==label, feature],
             label=iris_data.target_names[label],
             color=color)
    plt.xlabel(iris_data.feature_names[feature])
    plt.legend()
plt.show()
```

```python
df_iris = pd.DataFrame(iris_data.data, columns=iris_data.feature_names)
df_iris['target'] = iris_data.target
df_iris.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(iris_data['data'], iris_data['target'], random_state=0)
```

```python
pd.Series(y_train).value_counts()
```

```python
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train, y_train)
```

```python
pred_y = knn.predict(X_test)
```

```python
print(f'Test score: {f1_score(pred_y, y_test, average="micro")}')
```
