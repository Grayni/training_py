```python
import matplotlib.pyplot as plt
%matplotlib inline

import numpy as np
import pandas as pd
import seaborn as sns

from sklearn.preprocessing import StandardScaler
```

```python
# Generation of dataset
df = pd.DataFrame({
    'exponential': np.random.exponential(10, 1000),  # Экспоненциальное распределение, 10 – "резкость" экспоненты, 1000 – размер
    'normal_p': np.random.normal(10, 2, 1000)})  # Нормальное распределение, 10 – среднее значение р., 2 – стандартное отклонение, 1000 размер

fig, (ax1) = plt.subplots(ncols=1, figsize=(10, 8))
ax1.set_title('Original Distribution')

# kdeplot() (KDE – оценка плотности ядра) – специальный метод для графиков распределений
sns.histplot(df['exponential'], color='red')
sns.histplot(df['normal_p'], color='green')
plt.show()
```

```python
scaler = StandardScaler()

# create dataset scaler
arr_scaler = scaler.fit_transform(df)

# copy name of cols
names = list(df.columns)

# scaler in DataFrame
df_scaler = pd.DataFrame(arr_scaler, columns=names)

fig, (ax1) = plt.subplots(ncols=1, figsize=(10, 8))

ax1.set_title('after Standirtization')

sns.histplot(df_scaler['exponential'], color='red')
sns.histplot(df_scaler['normal_p'], color='green')
plt.show()
```
