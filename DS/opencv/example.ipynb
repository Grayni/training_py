```python
import cv2
import matplotlib.pyplot as plt
plt.style.use('dark_background')
import numpy as np
```

```python
plt.rcParams['figure.figsize'] = (7, 7)
```

```python
img = cv2.imread('images/station.jpg', cv2.IMREAD_COLOR)
plt.imshow(img)
```

```python
rgb_img = img[...,::-1]
plt.imshow(rgb_img)
```

```python
gray_img = cv2.imread('images/station.jpg', cv2.IMREAD_GRAYSCALE)
plt.imshow(gray_img, cmap='gray')
```

```python
img_1 = cv2.medianBlur(img, 3)
```

```python
plt.imshow(img_1)
```

```python
y = cv2.Sobel(gray_img, cv2.CV_16S, 0, 1)
abs_y = cv2.convertScaleAbs(y)
plt.imshow(abs_y)
```

```python
def rotate(img, angle):
    h, w = img.shape[:2]
    center = (w / 2, h / 2)
    mtx = cv2.getRotationMatrix2D(center, angle, scale=1.0)
    return cv2.warpAffine(img, mtx, (w, h))
```

```python
plt.imshow(rotate(gray_img, -45), cmap='gray')
```
