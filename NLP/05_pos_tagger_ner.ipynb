```python
import nltk
# nltk.download('punkt')
# nltk.download('averaged_perceptron_tagger_ru')
import requests
from bs4 import BeautifulSoup
import re

def url_to_string(url):
    res = requests.get(url)
    html = res.text
    soup = BeautifulSoup(html, features='html5lib')
    for script in soup(["script", "style", 'aside']):
        script.extract()
    return " ".join(re.split(r'[\n\t]+', soup.get_text()))

document = url_to_string('https://gorky.media/context/ne-tolko-kutzee/')

doc = nltk.pos_tag(nltk.word_tokenize(document, language='russian'), lang='rus')

doc
```

```python
{(' '.join(c[0] for c in chunk), chunk.label() ) for chunk in nltk.ne_chunk(nltk.pos_tag(nltk.word_tokenize(document))) if hasattr(chunk, 'label') }
```

```python
from nltk.tag import DefaultTagger
from nltk.tag import UnigramTagger
from nltk.tag import BigramTagger, TrigramTagger
from nltk.tag import RegexpTagger
```

```python
# tags

tags = [tag for (word, tag) in doc]
nltk.FreqDist(tags)
```

```python
# sentences

from nltk.tokenize import sent_tokenize
sentences = sent_tokenize(document, language='russian')

sentences_tags = []

for sent in sentences:
    sentences_tags.append(nltk.pos_tag(nltk.word_tokenize(sent, language='russian'), lang='rus'))

sentences_tags
```

```python
from nltk.tokenize import wordpunct_tokenize

biography_text = 'Фёдор Михайлович Достоевский, родившийся в 1821 году в Москве, стал одним из самых выдающихся и влиятельных писателей в русской и мировой литературе.'

train_data = sentences_tags[:int(len(sentences_tags) * 0.9)]
test_data = sentences_tags[int(len(sentences_tags) * 0.9):]
test_sent = wordpunct_tokenize(biography_text)

test_data[0]
```

```python
# DefaultTagger

default_tagger = DefaultTagger('S')
display(default_tagger.tag(test_sent), default_tagger.accuracy(test_data))
```

```python
# UnigramTagger

unigram_tagger = UnigramTagger(train_data)
display(unigram_tagger.tag(test_sent), unigram_tagger.accuracy(test_data))
```

```python
# BigramTagger

bigram_tagger = BigramTagger(train_data, backoff=unigram_tagger)
display(bigram_tagger.tag(test_sent), bigram_tagger.accuracy(test_data))
```

```python
# TrigramTagger

trigram_tagger = TrigramTagger(train_data, backoff=bigram_tagger)
display(trigram_tagger.tag(test_sent), trigram_tagger.accuracy(test_data))
```

```python
# RegexpTagger

import string

punctuations = string.punctuation
pronouns = ['я', 'мы', 'ты', 'вы', 'он', 'она', 'оно', 'они', 'себя', 'свой']
prepositions = ['в', 'на', 'по', 'за', 'до', 'из', 'к', 'от', 'через', 'у', 'о', 'об', 'без', 'вроде', 'включая', 'исключая' ]

patterns = [
    (r'^[a-zA-Z]+$', 'NONLEX'),
    (r'^\d+$', 'NUM=ciph'), # Цифры (например, '26', '2017')
    (r'^\d+-\d+$', 'NUM=ciph'), # Диапазоны чисел (например, '1967-1970')
    (r'^(' + '|'.join(prepositions) + ')$', 'PR'),
    (r'^(' + '|'.join(pronouns) + ')$', 'S-PRO'),
    (r'[' + punctuations + r']', 'NONLEX'),

    (r'.*ить$', 'V'), # глаголы
    (r'.*ся$', 'V'), # возвратные глаголы

    (r'.*ость$', 'S'), # существительные от прилагательных

    (r'^[А-ЯЁ].*$', 'S'), # существительные с заглавной буквы

    (r'.*но$', 'ADV-PRO'), # наречие

    (r'^\d+$', 'NUM'), # числа

    (r'.*', 'S'), # существительные по умолчанию

      # глаголы
    (r'.*иться$', 'V'),
    (r'.*ить$', 'V'),
    (r'.*ать$', 'V'),
    (r'.*ять$', 'V'),
    (r'.*еть$', 'V'),
    (r'.*нуть$', 'V'),
    (r'.*вать$', 'V'),

    # существительные

    (r'.*ость$', 'S'),
    (r'.*ание$', 'S'),
    (r'.*ение$', 'S'),
    (r'.*тель$', 'S'),
    (r'.*ние$', 'S'),

    # прилагательные

    (r'.*ый$', 'A'),
    (r'.*ий$', 'A'),
    (r'.*ая$', 'A'),
    (r'.*яя$', 'A'),

    # наречия

    (r'.*о$', 'ADV'),

    # порядковые числительные

    (r'.*ое$', 'ANUM'),
    (r'.*ая$', 'ANUM'),

    # притяжательные местоимения

    (r'.*ий$', 'A-PRO'),
    (r'.*я$', 'A-PRO'),

    # краткие прилагательные

    (r'.*ая$', 'A=f'),
    (r'.*ое$', 'A=n'),
    (r'.*ие$', 'A=pl'),

    # префиксы

    (r'не.*', 'ADV-PRO'), # не-
    (r'без.*', 'PR'),
    (r'.*ая$', 'A-PRO=f'), # женский род прилагательных
    (r'.*ий$', 'A-PRO=sg'), # мужской род прилагательных
    (r'.*ые$', 'A-PRO=pl'), # множественное число прилагательных
]
regexp_tagger = RegexpTagger(patterns)
display(regexp_tagger.tag(test_sent), regexp_tagger.accuracy(test_data))
```
