```python
import numpy as np
from sklearn.datasets import make_regression
```

```python
np.random.seed(9)
X_rgr, y_rgr = make_regression(n_samples=1000, n_features=2, n_informative=1, n_targets=1)

display(X_rgr, y_rgr)
```

```python
def corr(x, y):
    x_mean = x.mean()
    y_mean = y.mean()
    regression = np.sum((x - x_mean) * (y - y_mean)) / np.sqrt(np.sum((x - x_mean)**2) * np.sum((y - y_mean)**2))
    return regression
```

```python
print(f'corr 0: {corr(X_rgr[:, 0], y_rgr)}')
print(f'corr 1: {corr(X_rgr[:, 1], y_rgr)}')
```

```python
# np.corrcoef
np.corrcoef(X_rgr[:, 0], y_rgr)
```

```python
print(f'corr 0: {np.corrcoef(X_rgr[:, 0], y_rgr)[0][1]}')
print(f'corr 1: {np.corrcoef(X_rgr[:, 1], y_rgr)[0][1]}')
```

```python
# PCA
np.random.seed(7)
x = np.arange(1, 11)
y = 2 * x + np.random.randn(10) * 2

X = np.vstack((x, y)).T
plt.scatter(X[:, 0], X[:, 1])
plt.show()
```

```python
def standard_scale(x):
    res = (x - x.mean(axis=0)) / x.std(axis=0)
    return res

X = standard_scale(X)
plt.scatter(X[:, 0], X[:, 1])
plt.grid()
plt.show()
```

```python
def covariance(x, y):
    return np.sum(x * y) / (len(x) - 1)

print(f'covariance 0: {covariance(X[:, 0], X[:, 0])}')
print(f'covariance 1: {covariance(X[:, 1], X[:, 1])}')
print(f'covariance 0 & 1: {covariance(X[:, 0], X[:, 1])}')
```

```python
# Example iris
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
X = iris.data
y = iris.target
display(X, y)
```

```python
X = X.astype(float)
X = standard_scale(X)
```

```python
X[:5]
```

```python
covariance_matrix = X.T @ X

eig_values, eig_vectors = np.linalg.eig(covariance_matrix)
eig_pairs = [(np.abs(eig_values[i]), eig_vectors[:, i]) for i in range(len(eig_values))]
eig_pairs.sort(key=lambda x: x[0], reverse=True)

print('own values & own vectors in descending order:')

for i in eig_pairs:
    print(i)
```

```python
eig_sum = sum(eig_values)

percentage_eig = [pair / eig_sum * 100 for pair in sorted(eig_values, reverse=True)]
percentage_eig
```

```python
cum_var_exp = np.cumsum(percentage_eig)
cum_var_exp
```

```python
W = np.hstack([eig_pairs[i][1].reshape(4, 1) for i in range(2)])
W
```

```python
Z = X.dot(W)
Z[:5]
```

```python
plt.figure()

for c, i in zip('rgb', [0, 1, 2]):
    plt.scatter(Z[y == i, 0], Z[y==i, 1], c=c)

plt.xlabel('Component 1')
plt.ylabel('Component 2')
plt.title('PCA')
plt.grid()
plt.show()
```
