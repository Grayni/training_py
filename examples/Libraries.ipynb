#### Scipy

```python
from scipy import integrate
```

```python
# Integral
def sinx(x):
    return np.sin(x) * x**2

integrate.quad(sinx, -3, 10)
```

```python
# Determinant
scipy.linalg.det(arr_scipy)
```

```python
# Inverse matrix
scipy.linalg.inv(arr_scipy)
```

```python
# eigenvalues of the matrix
scipy.linalg.eig(arr_scipy)
```

#### Matplotlib

```python
fig = plt.figure(figsize=(16, 4))
x = np.arange(-2 * np.pi, 2 * np.pi, 0.1)
plt.plot(x, np.sin(x))
plt.xlabel('x')
plt.ylabel('y')
plt.title(label='sin')
plt.grid()
plt.show()
```

```python
fig = plt.figure(figsize=(16, 2))
plt.subplot(1, 2, 1)
plt.plot(x, np.sin(x), color='red')
plt.subplot(1, 2, 2)
plt.plot(x, np.cos(x), color='green')
plt.show()
```

```python
fig = plt.figure(figsize=(16, 2))
plt.subplot(1, 1, 1)
plt.plot(x, np.sin(x), color='red')
plt.subplot(1, 1, 1)
plt.plot(x, np.cos(x), color='green')
plt.show()
```
