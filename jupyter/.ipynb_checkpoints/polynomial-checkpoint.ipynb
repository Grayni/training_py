```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
inp = '6'  # value of polynomials

inp_arr = ['0 1', '1 1', '2 25', '-1 1', '-2 -23', '0.5 0.90625']  # [[x y], [x y] ...]
M = np.array([])
v = np.array([])


for i in range(int(inp)):
    empty = np.array([])
    v = np.append(v, float(inp_arr[i].split()[1]))

    for j in range(int(inp)):
        empty = np.append(empty, float(inp_arr[i].split()[0]) ** j)

    if M.shape[0] == 0:
        M = np.array([empty])
    else:
        M = np.concatenate((M, [empty]))

c = np.linalg.solve(M, v)

def f(x, c):
    return c[0]+c[1]*x+c[2]*x**2+c[3]*x**3+c[4]*x**4 + c[5]*x**5

x = np.arange(-1.2, 1, 0.01)
plt.plot(x, f(x, c))
plt.xlabel(r'$x$')
plt.ylabel(r'$f(x)$')
plt.grid(True, color='pink')
plt.show()
```
