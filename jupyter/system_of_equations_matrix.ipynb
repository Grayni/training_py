$$ A⋅X = B $$

```python
A = np.array([[3, 2], [3, -4]])
B = np.array([4, 1])
np.linalg.solve(A, B)
```

$$X = A^{-1}⋅B$$

```python
A1 = np.linalg.inv(A)
A1
```

```python
print('det = ', np.linalg.det(A))
np.dot(A1, B)
```
