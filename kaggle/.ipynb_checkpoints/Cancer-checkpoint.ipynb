```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
data = pd.read_csv('input/cancer/survey_lung_cancer.csv', sep=',', header=0)
```

```python
data.head()
```

```python
data.info()
```

```python
data.describe(include=['O'])
```

```python
from sklearn.preprocessing import LabelEncoder

lec = LabelEncoder()
#data['GENDER'] = data['GENDER'].replace({'M': 'Male', 'F': 'Female'})
data['LUNG_CANCER'] = lec.fit_transform(data['LUNG_CANCER'])


data = pd.get_dummies(data, columns=['GENDER'])
data.rename(columns={'GENDER_M': 'MALE', 'GENDER_F': 'FEMALE'}, inplace=True)
```

```python
data.describe()
```

```python
features_redact = ['SMOKING', 'YELLOW_FINGERS', 'ANXIETY', 'PEER_PRESSURE',
       'CHRONIC DISEASE', 'FATIGUE ', 'ALLERGY ', 'WHEEZING',
       'ALCOHOL CONSUMING', 'COUGHING', 'SHORTNESS OF BREATH',
       'SWALLOWING DIFFICULTY', 'CHEST PAIN']

data[features_redact] = data[features_redact].apply(lambda x: x.replace({1: 0, 2: 1}))
```

#### Correlation matrix

```python
corr = data.corr()
mask = np.zeros_like(corr, dtype=bool)
mask[np.triu_indices_from(mask)] = True

f, ax = plt.subplots(figsize=(16, 8))
cmap = sns.diverging_palette(220, 10, as_cmap=True)
sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.5, center=0, square=True, linewidths=.5, cbar_kws={'shrink': .5}, cbar=True, fmt='.2f', annot_kws={'size': 10}, annot=True)
plt.show()
```

```python
# Unique values in features
data.apply(lambda feat: len(feat.unique())) 
```

```python
# Categorical Age
data['CATEGORICAL_AGE'] = pd.cut(data['AGE'], 3, precision=0)
data[['CATEGORICAL_AGE', 'LUNG_CANCER']].groupby('CATEGORICAL_AGE').mean().sort_values(by='CATEGORICAL_AGE', ascending=False)

data['CATEGORICAL_AGE'] = pd.to_numeric(data['CATEGORICAL_AGE'].apply(lambda x: x.right)).astype('int64')
```

```python
data.head()
```

```python
y = data['LUNG_CANCER']
X = data.drop(['LUNG_CANCER', 'SMOKING', 'SHORTNESS OF BREATH'], axis=1)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X = scaler.fit_transform(X)
```

#### Models

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score


classifiers = [
    KNeighborsClassifier(5),
    SVC(probability=True, kernel='rbf'),
    DecisionTreeClassifier(),
    RandomForestClassifier(),
    GaussianNB(),
    LogisticRegression(solver='lbfgs', max_iter=500)
]

X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size=0.23, random_state=42)

acc_dict = {}
```

```python
for clf in classifiers:
    name =clf.__class__.__name__
    clf.fit(X_train, y_train)
    train_pred = clf.predict(X_valid)
    acc = accuracy_score(y_valid, train_pred)
    
    if name in acc_dict:
        acc_dict[name] +=acc
    else:
        acc_dict[name] = acc
```

```python
log_cols = ['Classifier', 'Accuracy']
log = pd.DataFrame(columns=log_cols)

for clf in acc_dict:
    acc_dict[clf] = acc_dict[clf] / 10
    log_entry = pd.DataFrame([[clf, acc_dict[clf]]], columns=log_cols)
    log = pd.concat([log, log_entry])
```

```python
plt.xlabel('Accuracy')
plt.title('Classifier Accuracy')

sns.set_color_codes('muted')
sns.barplot(x='Accuracy', y='Classifier', data=log.sort_values(by='Accuracy'), color='b')
plt.show()
```

```python
for clf in classifiers:
    name =clf.__class__.__name__
    clf.fit(X_train, y_train)
    train_pred = clf.predict(X_valid)
    acc = f1_score(y_valid, train_pred)
    
    if name in acc_dict:
        acc_dict[name] +=acc
    else:
        acc_dict[name] = acc
```

```python
log_cols = ['Classifier', 'f1']
log = pd.DataFrame(columns=log_cols)

for clf in acc_dict:
    acc_dict[clf] = acc_dict[clf] / 10
    log_entry = pd.DataFrame([[clf, acc_dict[clf]]], columns=log_cols)
    log = pd.concat([log, log_entry])
```

```python
plt.xlabel('f1')
plt.title('Classifier f1')

sns.set_color_codes('muted')
sns.barplot(x='f1', y='Classifier', data=log.sort_values(by='f1'), color='b')
plt.show()
```

#### LogisticRegression

```python
from sklearn.metrics import classification_report

lr = LogisticRegression()
lr.fit(X_train, y_train)
lr_pred = lr.predict(X_valid)
lr_report = classification_report(y_valid, lr_pred, zero_division=0)
lr_acc = round(accuracy_score(y_valid, lr_pred)*100, ndigits=3)
print(f'\nClassification Report : \n\n{lr_report}')
print(f'\nThe Accuracy of Logistic Regression is {lr_acc} %')
```

```python
f1_score(y_valid, lr_pred, zero_division=0)  # LogisticRegression
```

#### RandomForestClassifier

```python
rf = RandomForestClassifier()

rf.fit(X_train, y_train)
rf_pred = rf.predict(X_valid)
rf_report = classification_report(y_valid, rf_pred, zero_division=0)
rf_acc = round(accuracy_score(y_valid, rf_pred)*100, ndigits=3)
print(f'\nClassification Report : \n\n{rf_report}')
print(f'\nThe Accuracy of Logistic Regression is {rf_acc} %')
```
