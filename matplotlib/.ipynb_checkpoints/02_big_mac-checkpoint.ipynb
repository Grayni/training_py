```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt

%matplotlib inline
```

#### Визуализация данных на примере Биг-Мак

Индекс Биг-Мака - неофициальный способ определения паритета покупательской способности.

```python
# Данные с сайта журнала The Economist за январь 2018 года
url = 'http://infographics.economist.com/2018/databank/BMFile2000toJan2018.xls'
```

```python
# С помощью функции read_excel можно читать данные из локального или удаленного источников
bmi = pd.read_excel(url)
```

```python
bmi.shape
```

```python
bmi.columns
```

```python
bmi.iloc[:, :4].head(10)
```

```python
bmi_price_series = bmi.set_index('Country')['dollar_price']
bmi_price_series.head()
```

```python
bmi_price_series.plot(kind='barh')
plt.show()
```

```python
from pylab import rcParams
rcParams['figure.figsize'] = 9, 12
```

```python
bmi_price_series.plot(kind='barh')
plt.show()
```

```python
bmi_price_series.sort_values().plot(kind='barh')
plt.show()
```

```python
bmi_price_series[bmi_price_series.index == 'Switzerland'][0]
```

```python
bmi_price_series.sort_index().plot(kind='barh')
plt.show()
```

```python
bmi_price_series.sort_index(ascending=False).plot(kind='barh')
plt.show()
```

```python
rcParams['figure.figsize'] = 4.5, 6
plt.hist(bmi_price_series, ec='white', color='lightgray')
plt.show()
```

```python
# Сохранение графиков в файл
plt.hist(bmi_price_series, ec='white', color='lightgray')
plt.savefig('./img/BM_price.png')
plt.show()
```
