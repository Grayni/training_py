```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

```python
x = np.linspace(-5, 5, 101)
x2 =x**2
x3 = x**3
```

```python
plt.axis([-5, 5, -8, 8])
plt.plot(x, x2)
plt.plot(x, x3)
```

```python
fig, ax = plt.subplots(nrows=1, ncols=1)
ax.plot(x, x2)
```

```python
# Создание объектов figure и axes
fig, ax = plt.subplots(nrows=1, ncols=2)
ax1, ax2 = ax
ax1.plot(x, x2)
ax2.plot(x, x3)
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2)
ax1, ax2 = ax.flatten()
ax1.plot(x, x2)
ax2.plot(x, x3)
```

```python
# Свойства объекта figure
fig, ax = plt.subplots(nrows=1, ncols=2)
ax1, ax2 = ax.flatten()
ax1.plot(x, x2)
ax2.plot(x, x3)
fig.set_size_inches(8, 3.5)
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2)
ax1, ax2 = ax.flatten()
ax1.plot(x, x2)
ax2.plot(x, x3)
fig.set_size_inches(8, 3.5)
plt.subplots_adjust(wspace=0.3)
```

```python
fig, ax = plt.subplots(nrows=2, ncols=1)
ax1, ax2 = ax.flatten()
ax1.plot(x, x2)
ax2.plot(x, x3)
fig.set_size_inches(3, 6)
```

```python
fig, ax = plt.subplots(nrows=2, ncols=1)
ax1, ax2 = ax.flatten()
ax1.plot(x, x2)
ax2.plot(x, x3)
fig.set_size_inches(3, 6)
plt.subplots_adjust(hspace=0.25)
```

```python
y = np.sin(x)
z = np.cos(x)
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2)
ax1, ax2, ax3, ax4 = ax.flatten()
ax1.plot(x, x2)
ax2.plot(x, x3)
ax3.plot(x, y)
ax4.plot(x, z)
fig.set_size_inches(8, 6)
plt.subplots_adjust(wspace=0.3, hspace=0.3)
```
