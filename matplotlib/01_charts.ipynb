### Логарифмический масштаб

```python
planets = ['Mercury', 'Venus', 'Earth', 'Mars', 'Jupiter', 'Saturn', 'Uranus', 'Neptune']
masses = [0.055274, 0.815, 1.0, 0.107, 317.8, 95.0, 14.6, 17.147] # отношение масс планет к массе Земли
```

```python
plt.plot(planets, masses)
plt.show()
```

```python
masses[planets.index('Earth')] / masses[planets.index('Mercury')] 
```

```python
# Логарифмический масштаб на вертикальной оси
plt.plot(planets, masses)
plt.yscale(value='log')
plt.show()
```

```python
planet_info = pd.Series(masses, index=planets)
planet_info
```

```python
planet_info.plot()
plt.show()
```

```python
my_colors = ['gold', 'red', 'green', 'blue', 'cyan', 'pink', 'brown', 'olive']
```

```python
# Столбчатая диаграмма
planet_info.plot(kind='bar', color=my_colors)
plt.show()
```

```python
planet_info.plot(kind='bar', logy=True, color=my_colors)
plt.show()
```

```python
planet_info.plot(kind='barh', logx=True, color=my_colors)
plt.show()
```

### Построение гистограмм

```python
a = [i for i in range(10)]
a
```

```python
hist_info = plt.hist(a, bins=3)
print(hist_info)
plt.show()
```

```python
plt.hist(a, bins=3, edgecolor='white')
plt.show()
```

```python
plt.hist(a, bins=3, ec='white', orientation='horizontal')
```
