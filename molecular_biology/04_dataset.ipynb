```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import tarfile
```

#### Unzip data

```python
archive_path = 'databases/clinical.project-tcga-dlbc.2023-09-20.tar.gz'
extract_path = 'databases/clinical'

with tarfile.open(archive_path, 'r:gz') as tar:
    tar.extractall(extract_path)
```

```python
# func: load pandas data
def table(tb_name):
    tb = pd.read_csv(f'{extract_path}/{tb_name}', sep='\t')
    display(tb.head())
    return tb
```

#### Tables refactoring

```python
# help func: show variable name
def print_variable_name(var):
    for name, value in globals().items():
        if value is var:
            return name

# remove empty cols
value_to_check = "'--"
before_clear = len(cl_data.columns)
cl_data = cl_data.drop(cl_data.columns[cl_data.eq(value_to_check).all()], axis=1)
f'Data: {print_variable_name(cl_data)} >>> before clear: {before_clear}, after clear: {len(cl_data.columns)}'
```

```python
# remove case_id
if 'case_id' in cl_data.columns:
    cl_data.drop('case_id', axis=1, inplace=True)
```

```python
# remove cols (uninformative)
target_values = ["0", "'--" , "not reported", "no", "Malignant lymphoma, large B-cell, diffuse, NOS", "9680/3", "No"]
columns_rm = set()

for col in cl_data:
    if len(cl_data[col].unique()) == 1:
        cl_data = cl_data.drop(col, axis=1)
    
    # remove cols containing set target_values
    elif len(cl_data[col].unique()) == 2:
        for unq in cl_data[col].unique():
            if not unq in target_values:
                columns_rm.add(col)
                
        if not col in columns_rm:
            cl_data = cl_data.drop(col, axis=1)
               
len(cl_data.columns)
```

```python
# identical columns
if 'tissue_or_organ_of_origin' in cl_data.columns:
    comparison_result = cl_data['site_of_resection_or_biopsy'] == cl_data['tissue_or_organ_of_origin']
    print(comparison_result.prod()) # if 1 -> identical
    
    cl_data.drop('tissue_or_organ_of_origin', axis=1, errors='ignore', inplace=True)
    
len(cl_data.columns)
```

```python
# category
# category = ['ethnicity', 'gender', 'race', 'vital_status', 'ann_arbor_b_symptoms', 'ann_arbor_clinical_stage', 'ann_arbor_extranodal_involvement', 'icd_10_code', 'prior_treatment', 'treatment_or_therapy', 'treatment_type']

# numbers int
# cols_int = ['age_at_index', 'days_to_birth', 'days_to_death', 'year_of_birth', 'year_of_death', 'age_at_diagnosis', 'days_to_last_follow_up', 'year_of_diagnosis']

# cols with strings
# cols_str = ['case_submitter_id', 'site_of_resection_or_biopsy']

# '-- to NaN 
cl_data = cl_data.replace("'--", pd.NA)
```

```python
# category
category = ['ethnicity', 'gender', 'race', 'vital_status', 'ann_arbor_b_symptoms', 'ann_arbor_clinical_stage', 'ann_arbor_extranodal_involvement', 'icd_10_code', 'prior_treatment', 'treatment_or_therapy', 'treatment_type']

cl_data[category] = cl_data[category].apply(lambda x: pd.factorize(x)[0])
cl_data[category] = cl_data[category].astype('category')
cl_data.head()
```

```python
# numbers int
cols_int = ['age_at_index', 'days_to_birth', 'days_to_death', 'year_of_birth', 'year_of_death', 'age_at_diagnosis', 'days_to_last_follow_up', 'year_of_diagnosis']

for col in cols_int:
    cl_data[col] = pd.to_numeric(cl_data[col], errors='coerce')
    cl_data[col] = cl_data[col].fillna(cl_data[col].median())

cl_data[cols_int] = cl_data[cols_int].astype('int64')
cl_data.head()
```

```python
# strings
cols_str = ['case_submitter_id', 'site_of_resection_or_biopsy']
cl_data[cols_str] = cl_data[cols_str].apply(lambda x: pd.factorize(x)[0])
```

#### Correlation

```python
correlation_matrix = cl_data.corr()
correlation_matrix
```

```python
# Mask triangle
mask = np.triu(np.ones_like(correlation_matrix, dtype=bool))

# Style
sns.set(style="white")

# corr matrix
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, mask=mask, annot=True, cmap='coolwarm', fmt=".2f", annot_kws={"fontsize": 8})
plt.title('Корреляционная матрица')
plt.show()
```

```python
# if correlation index > 0.8 remove col
np.fill_diagonal(correlation_matrix.values, np.nan)
high_corr_values = correlation_matrix[(correlation_matrix > 0.8) | (correlation_matrix < -0.8)]

high_corr_values = high_corr_values.stack()
print(high_corr_values)
```
