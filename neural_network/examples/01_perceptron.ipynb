```python
import numpy as np
import matplotlib.pyplot as plt
```

### w*x

```python
N = 5
```

```python
x1 = np.random.random(N)
x2 = x1 + [np.random.randint(10) / 10 for _ in range(N)]
C1 = [x1, x2]

x1 = np.random.random(N)
x2 = x1 - [np.random.randint(10) / 10 for _ in range(N)] - 0.1
C2 = [x1, x2]
```

```python
f = [0, 1]
```

```python
w = np.array([-0.3, 0.3])
for i in range(N):
    x = np.array([C2[0][i], C2[1][i]])
    y = np.dot(w, x)
    if y >= 0 :
        print('Class C1')
    else:
        print('Class C2')
```

```python
plt.scatter(C1[0][:], C1[1][:], s=10, c='r')
plt.scatter(C2[0][:], C2[1][:], s=10, c='g')
plt.plot(f)
plt.grid()
plt.show()
```

### w*x+b

```python
N = 5
b = 3
```

```python
x1 = np.random.random(N)
x2 = x1 + [np.random.randint(10) / 10 for _ in range(N)] + b
C1 = [x1, x2]
```

```python
x1 = np.random.random(N)
x2 = x1 - [np.random.randint(10) / 10 for _ in range(N)] - 0.1 + b
C2 = [x1, x2]
```

```python
f = [0+b, 1+b]
```

```python
w2 = 0.5
w3 = -b*w2
w = np.array([-w2, w2, w3])
for i in range(N):
    x = np.array([C1[0][i], C1[1][i], 1])
    y = np.dot(w, x)
    if y >= 0:
        print('Class C1')
    else:
        print('Class C2')
```

```python
plt.scatter(C1[0][:], C1[1][:], c='r', s=10)
plt.scatter(C2[0][:], C2[1][:], c='g', s=10)
plt.plot(f)
plt.grid()
plt.show()
```

### XOR

```python
def act(x):
    return 0 if x <= 0 else 1
```

```python
def go(C):
    x = np.array([C[0], C[1], 1])
    w1 = [1, 1, -1.5]
    w2 = [1, 1, -0.5]
    w_hidden = np.array([w1, w2])
    w_out = np.array([-1, 1, -0.5])

    sum = np.dot(w_hidden, x)
    out = [act(x) for x in sum]
    out.append(1)
    out = np.array(out)

    sum = np.dot(w_out, out)
    y = act(sum)
    return y
```

```python
C1 = [(1, 0), (0, 1)]
C2 = [(0, 0), (1, 1)]
```

```python
print(go(C1[0]), go(C1[1]))
print(go(C2[0]), go(C2[1]))
```
