```python
import cv2
```

```python
cap = cv2.VideoCapture(0)
fourcc = cv2.VideoWriter_fourcc(*'XVID')
out = cv2.VideoWriter('video/output.avi', fourcc, 20.0, (640, 480))
```

```python
print(cap.isOpened())

# True if cap exist
while cap.isOpened():
    # if ret == True, camera ON and frame capture
    ret, frame = cap.read()

    if ret:
        # show width and height of camera
        print(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
        print(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

        out.write(frame)

        gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

        # display frame in gray
        cv2.imshow('frame', gray)
        # display frame in color
        # cv2.imshow('frame', frame)

        # click on q -> exit cycle (camera OFF)
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
    else:
        break

# release camera
cap.release()
out.release()
cv2.destroyAllWindows()
```
