```python
from scipy.misc import derivative
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-1, 6, 100)
```

```python
def func(x):
    return x**2 * np.exp(-x)
```

```python
def derivate(x):
    return derivative(func, x)
```

```python
plt.plot(x, derivate(x))
plt.show()
```

```python
derivate(-1)
```

```python
plt.plot(x, np.array([func(val) for val in x]))
plt.show()
```
