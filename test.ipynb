```python
## np.squeeze vs np.ravel vs np.flatten

test_ar = np.array([[[1], [2]], [[2], [3]], [[3], [4]]])

# np.squeeze  # remove all or last "nested" axis, return copy
squeeze1 = np.squeeze(test_ar, axis=2)  # or axis=None
print(squeeze1)
squeeze2 = np.squeeze([[[1]], [[2]], [[3]]], axis=None)
print(squeeze2)

# np.ravel create copy of one-dimensional array
ravel = np.ravel(test_ar)
print(ravel)

# arr.flatten  # method of np.arr to one-dimensional array
test_ar = test_ar.flatten()
print(test_ar)
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def adjust_curves(image, curves):
    # Create a lookup table for the curves
    num_channels = len(curves)
    lut = np.zeros((256, num_channels), dtype=np.uint8)
    for j in range(num_channels):
        for i in range(256):
            lut[i, j] = np.clip(curves[j][0][i] * i + curves[j][1][i], 0, 255)

    # Apply the lookup table to the image
    adjusted_image = cv2.LUT(image, lut)

    return adjusted_image

# Load an image
image = cv2.imread('pictures/android.png')

# Define the curves
curves = [
    [[0.5, 1, 1]],  # Red channel
    [[0, 0.5, 1], [0.25, 0.5, 0.75]],  # Green channel
    [[0, 0, 0.5], [0.25, 0.25, 0.5], [0.75, 0.5, 1]]   # Blue channel
]

# Adjust the curves of the image
adjusted_image = adjust_curves(image, curves)

# Display the original and adjusted images
plt.subplot(121), plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB)), plt.title('Original')
plt.subplot(122), plt.imshow(cv2.cvtColor(adjusted_image, cv2.COLOR_BGR2RGB)), plt.title('Adjusted')
plt.show()
```
